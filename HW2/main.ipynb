Consider the vector space 
${ð‘_2}^ð‘›$, where each vector in this space corresponds to a binary string of length ð‘›. The addition of two vectors in this space is equivalent to the XOR operation between two binary strings. Additionally, a linear combination of some members of this space corresponds to the XOR of a subset of those vectors. The vectors $ð‘£_1, ð‘£_2, ..., ð‘£_ð‘š$ are given.
For each $1 â‰¤ ð‘˜ â‰¤ ð‘š$, check whether $ð‘£_ð‘˜ âˆˆ span(ð‘£_1, ð‘£_2, ..., ð‘£_{ð‘˜âˆ’1})$ i.e., whether vector $ð‘£_ð‘˜$ can be written as a linear combination of the previous vectors using XOR operations.

**Input:** The first line of input contains two integers, 
ð‘› and ð‘š, separated by a space. In each of the next ð‘š lines, there is a binary string of length ð‘›, without spaces between the digits.
$$1 â‰¤ ð‘›, ð‘š â‰¤ 500$$

**Output:** Print ð‘š lines of output, where each line contains 'YES' if the condition is satisfied for that vector, and 'NO' otherwise.

```python
def add_to_basis(number, n, base):
    for i in range(n):
        if (number >> i) & 1:
            if not base[i]:
                base[i] = number
                return False
            number ^= base[i]
    return True

def span_check(test_input):
    test_case = test_input.split()
    n = int(test_case[0])
    m = int(test_case[1])
    base = [None] * n
    vectors = [int(test_case[i + 2], 2) for i in range(m)]
    
    for vec in vectors:
        if add_to_basis(vec, n, base):
            print("YES")
        else:
            print("NO")
    print("################################3")
```
